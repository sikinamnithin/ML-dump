```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from tqdm import tqdm
import os
```

```python
np.random.seed(42)
torch.manual_seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])  # [-1, 1]
])

# Download CIFAR-10 (only 170MB, downloads quickly!)
print("Loading CIFAR-10 dataset...")
train_dataset = datasets.CIFAR10(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

train_loader = DataLoader(
    train_dataset,
    batch_size=128,
    shuffle=True,
    num_workers=4,
    pin_memory=True
)

print(f"Training samples: {len(train_dataset)}")
print(f"Image shape: {train_dataset[0][0].shape}")  # (3, 32, 32)
print(f"Classes: {classes}")

# Visualize real images
def show_images(images, labels=None, title="Images"):
    fig, axes = plt.subplots(2, 8, figsize=(16, 4))
    for i, ax in enumerate(axes.flatten()):
        if i < len(images):
            img = images[i]
            # Denormalize
            img = (img + 1) / 2
            img = img.permute(1, 2, 0).cpu().numpy()
            img = np.clip(img, 0, 1)
            ax.imshow(img)
            if labels is not None:
                ax.set_title(f"{classes[labels[i]]}")
        ax.axis('off')
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

# Show real images
sample_images = [train_dataset[i][0] for i in range(16)]
sample_labels = [train_dataset[i][1] for i in range(16)]
show_images(sample_images, sample_labels, "Real CIFAR-10 Images")
```

```python
class SinusoidalPositionEmbedding(nn.Module):
    """Sinusoidal time/noise level embedding"""
    def __init__(self, dim):
        super().__init__()
        self.dim = dim

    def forward(self, sigma):
        device = sigma.device
        half_dim = self.dim // 2
        emb = np.log(10000) / (half_dim - 1)
        emb = torch.exp(torch.arange(half_dim, device=device) * -emb)
        emb = sigma[:, None] * emb[None, :]
        emb = torch.cat([torch.sin(emb), torch.cos(emb)], dim=-1)
        return emb


class ResidualBlock(nn.Module):
    """Residual block with time embedding and group normalization"""
    def __init__(self, in_ch, out_ch, time_dim):
        super().__init__()
        self.time_mlp = nn.Sequential(
            nn.SiLU(),
            nn.Linear(time_dim, out_ch)
        )
        
        self.block1 = nn.Sequential(
            nn.GroupNorm(8, in_ch),
            nn.SiLU(),
            nn.Conv2d(in_ch, out_ch, 3, padding=1)
        )
        
        self.block2 = nn.Sequential(
            nn.GroupNorm(8, out_ch),
            nn.SiLU(),
            nn.Conv2d(out_ch, out_ch, 3, padding=1)
        )
        
        self.residual = nn.Conv2d(in_ch, out_ch, 1) if in_ch != out_ch else nn.Identity()
        
    def forward(self, x, t):
        h = self.block1(x)
        
        # Add time embedding
        time_emb = self.time_mlp(t)
        h = h + time_emb[:, :, None, None]
        
        h = self.block2(h)
        return h + self.residual(x)


class AttentionBlock(nn.Module):
    """Self-attention block for capturing long-range dependencies"""
    def __init__(self, channels):
        super().__init__()
        self.channels = channels
        self.norm = nn.GroupNorm(8, channels)
        self.qkv = nn.Conv2d(channels, channels * 3, 1)
        self.proj = nn.Conv2d(channels, channels, 1)
        
    def forward(self, x):
        B, C, H, W = x.shape
        h = self.norm(x)
        qkv = self.qkv(h)
        q, k, v = qkv.chunk(3, dim=1)
        
        # Reshape for attention
        q = q.view(B, C, H * W).transpose(1, 2)
        k = k.view(B, C, H * W)
        v = v.view(B, C, H * W).transpose(1, 2)
        
        # Attention
        scale = C ** -0.5
        attn = torch.softmax(torch.bmm(q, k) * scale, dim=-1)
        h = torch.bmm(attn, v)
        
        # Reshape back
        h = h.transpose(1, 2).view(B, C, H, W)
        return x + self.proj(h)


class DownBlock(nn.Module):
    """Downsampling block with residual blocks"""
    def __init__(self, in_ch, out_ch, time_dim, use_attention=False):
        super().__init__()
        self.res1 = ResidualBlock(in_ch, out_ch, time_dim)
        self.res2 = ResidualBlock(out_ch, out_ch, time_dim)
        self.attn = AttentionBlock(out_ch) if use_attention else nn.Identity()
        self.downsample = nn.Conv2d(out_ch, out_ch, 4, stride=2, padding=1)
        
    def forward(self, x, t):
        h = self.res1(x, t)
        h = self.res2(h, t)
        h = self.attn(h)
        return self.downsample(h), h  # Return downsampled and skip



class UpBlock(nn.Module):
    """Upsampling block with residual blocks - FIXED VERSION"""
    def __init__(self, in_ch, skip_ch, out_ch, time_dim, use_attention=False):
        super().__init__()
        self.upsample = nn.ConvTranspose2d(in_ch, in_ch, 4, stride=2, padding=1)
        # After concat, we have in_ch + skip_ch channels
        self.res1 = ResidualBlock(in_ch + skip_ch, out_ch, time_dim)
        self.res2 = ResidualBlock(out_ch, out_ch, time_dim)
        self.attn = AttentionBlock(out_ch) if use_attention else nn.Identity()
        
    def forward(self, x, skip, t):
        x = self.upsample(x)
        x = torch.cat([x, skip], dim=1)
        h = self.res1(x, t)
        h = self.res2(h, t)
        h = self.attn(h)
        return h


class ImprovedUNetScoreNetwork(nn.Module):
    """Advanced U-Net with attention for image generation - FIXED VERSION"""
    def __init__(self, in_channels=3, base_channels=128, time_dim=256):
        super().__init__()
        
        self.time_embed = nn.Sequential(
            SinusoidalPositionEmbedding(time_dim),
            nn.Linear(time_dim, time_dim * 4),
            nn.SiLU(),
            nn.Linear(time_dim * 4, time_dim)
        )
        
        # Initial projection
        self.init_conv = nn.Conv2d(in_channels, base_channels, 3, padding=1)
        
        # Encoder - each level outputs (downsampled, skip)
        # down1: 128 -> 128 (skip: 128 channels)
        self.down1 = DownBlock(base_channels, base_channels, time_dim)
        # down2: 128 -> 256 (skip: 256 channels)
        self.down2 = DownBlock(base_channels, base_channels * 2, time_dim)
        # down3: 256 -> 256 (skip: 256 channels)
        self.down3 = DownBlock(base_channels * 2, base_channels * 2, time_dim, use_attention=True)
        # down4: 256 -> 512 (skip: 512 channels)
        self.down4 = DownBlock(base_channels * 2, base_channels * 4, time_dim, use_attention=True)
        
        # Bottleneck with attention (512 channels)
        self.bottleneck = nn.Sequential(
            ResidualBlock(base_channels * 4, base_channels * 4, time_dim),
            AttentionBlock(base_channels * 4),
            ResidualBlock(base_channels * 4, base_channels * 4, time_dim)
        )
        
        # Decoder - FIXED: specify in_ch, skip_ch, out_ch correctly
        # up4: 512 (upsampled) + 512 (skip4) -> 256
        self.up4 = UpBlock(base_channels * 4, base_channels * 4, base_channels * 2, 
                          time_dim, use_attention=True)
        # up3: 256 (upsampled) + 256 (skip3) -> 256
        self.up3 = UpBlock(base_channels * 2, base_channels * 2, base_channels * 2, 
                          time_dim, use_attention=True)
        # up2: 256 (upsampled) + 256 (skip2) -> 128
        self.up2 = UpBlock(base_channels * 2, base_channels * 2, base_channels, time_dim)
        # up1: 128 (upsampled) + 128 (skip1) -> 128
        self.up1 = UpBlock(base_channels, base_channels, base_channels, time_dim)
        
        # Final output
        self.out = nn.Sequential(
            nn.GroupNorm(8, base_channels),
            nn.SiLU(),
            nn.Conv2d(base_channels, in_channels, 3, padding=1)
        )
        
    def forward(self, x, sigma):
        # Embed time/noise level
        t = self.time_embed(sigma.squeeze())
        
        # Initial conv
        x = self.init_conv(x)
        
        # Encoder
        x1, skip1 = self.down1(x, t)     # skip1: 128 channels
        x2, skip2 = self.down2(x1, t)    # skip2: 256 channels
        x3, skip3 = self.down3(x2, t)    # skip3: 256 channels
        x4, skip4 = self.down4(x3, t)    # skip4: 512 channels
        
        # Bottleneck
        x = self.bottleneck[0](x4, t)
        x = self.bottleneck[1](x)
        x = self.bottleneck[2](x, t)
        
        # Decoder
        x = self.up4(x, skip4, t)   # 512 + 512 -> 256
        x = self.up3(x, skip3, t)   # 256 + 256 -> 256
        x = self.up2(x, skip2, t)   # 256 + 256 -> 128
        x = self.up1(x, skip1, t)   # 128 + 128 -> 128
        
        return self.out(x)


# Reinitialize model with fixed architecture
model = ImprovedUNetScoreNetwork(
    in_channels=3,
    base_channels=128,
    time_dim=256
).to(device)

total_params = sum(p.numel() for p in model.parameters())
print(f"\nModel parameters: {total_params:,}")
print("Architecture fixed! Ready to train.")
```

```python
n_noise_levels = 10
sigma_min = 0.01
sigma_max = 1.0

sigmas = torch.exp(torch.linspace(np.log(sigma_max), np.log(sigma_min), n_noise_levels))
sigmas = sigmas.to(device)

print(f"Noise levels: {sigmas.cpu().numpy()}")
```

```python
class EMA:
    """Exponential Moving Average of model parameters"""
    def __init__(self, model, decay=0.9999):
        self.model = model
        self.decay = decay
        self.shadow = {}
        self.backup = {}
        
        for name, param in model.named_parameters():
            if param.requires_grad:
                self.shadow[name] = param.data.clone()
    
    def update(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                self.shadow[name] = self.decay * self.shadow[name] + (1 - self.decay) * param.data
    
    def apply_shadow(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                self.backup[name] = param.data
                param.data = self.shadow[name]
    
    def restore(self):
        for name, param in self.model.named_parameters():
            if param.requires_grad:
                param.data = self.backup[name]


def train_score_network(model, train_loader, sigmas, n_epochs=20, lr=0.0001):
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.01)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=n_epochs)
    ema = EMA(model, decay=0.9999)
    
    losses = []
    
    for epoch in range(n_epochs):
        model.train()
        epoch_loss = 0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{n_epochs}")
        
        for batch_idx, (x, _) in enumerate(pbar):
            x = x.to(device)
            batch_size = x.shape[0]
            
            # Randomly sample noise levels
            sigma_idx = torch.randint(0, len(sigmas), (batch_size,), device=device)
            sigma = sigmas[sigma_idx].view(-1, 1, 1, 1)
            
            # Add Gaussian noise
            z = torch.randn_like(x)
            x_noisy = x + sigma * z
            
            # Predict score
            score_pred = model(x_noisy, sigma.view(batch_size, 1))
            
            # Target score
            target = -z / sigma
            
            # Compute loss
            loss = torch.mean(
                sigma.squeeze() ** 2 * torch.sum((score_pred - target) ** 2, dim=(1, 2, 3))
            )
            
            # Optimize
            optimizer.zero_grad()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            
            # Update EMA
            ema.update()
            
            epoch_loss += loss.item()
            pbar.set_postfix({'loss': f'{loss.item():.4f}'})
            
            # Memory cleanup
            if batch_idx % 50 == 0:
                torch.cuda.empty_cache()
        
        avg_loss = epoch_loss / len(train_loader)
        losses.append(avg_loss)
        scheduler.step()
        
        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}, LR: {scheduler.get_last_lr()[0]:.6f}")
        
        # Generate samples using EMA model
        if (epoch + 1) % 5 == 0 or epoch == 0:
            ema.apply_shadow()
            sample_images(model, sigmas, n_samples=16, epoch=epoch+1)
            ema.restore()
    
    # Apply EMA for final model
    ema.apply_shadow()
    return losses
```

```python
@torch.no_grad()
def sample_langevin_dynamics(model, sigmas, n_samples=16, n_steps_per_sigma=100, 
                            step_size=0.00002, image_size=32):
    """Sample images using Langevin dynamics"""
    model.eval()
    
    # Start from random noise
    x = torch.randn(n_samples, 3, image_size, image_size).to(device) * sigmas[0]
    
    samples_progression = []
    
    for i, sigma in enumerate(tqdm(sigmas, desc="Sampling")):
        sigma_tensor = torch.ones(n_samples, 1).to(device) * sigma
        step_size_sigma = step_size * (sigma / sigmas[-1]) ** 2
        
        for step in range(n_steps_per_sigma):
            # Compute score
            score = model(x, sigma_tensor)
            
            # Langevin update
            noise = torch.randn_like(x)
            x = x + step_size_sigma * score + torch.sqrt(2 * step_size_sigma) * noise
            
            # Clip to valid range
            x = torch.clamp(x, -1, 1)
        
        samples_progression.append(x.cpu())
    
    return samples_progression


def sample_images(model, sigmas, n_samples=16, epoch=0, image_size=32):
    """Generate and display sample images"""
    samples_progression = sample_langevin_dynamics(
        model, sigmas, n_samples=n_samples, 
        n_steps_per_sigma=100, step_size=0.00002,
        image_size=image_size
    )
    
    # Visualize final samples
    final_samples = samples_progression[-1]
    fig, axes = plt.subplots(2, 8, figsize=(16, 4))
    for i, ax in enumerate(axes.flatten()):
        if i < len(final_samples):
            img = final_samples[i]
            # Denormalize from [-1, 1] to [0, 1]
            img = (img + 1) / 2
            img = img.permute(1, 2, 0).numpy()
            img = np.clip(img, 0, 1)
            ax.imshow(img)
        ax.axis('off')
    plt.suptitle(f"Generated Images (Epoch {epoch})")
    plt.tight_layout()
    plt.show()
```

```python
print("\n" + "="*80)
print("Starting Training on CelebA Faces...")
print("This will take a while! GPU recommended.")
print("="*80 + "\n")

# Train (increase epochs for better results)
losses = train_score_network(
    model, 
    train_loader, 
    sigmas, 
    n_epochs=120,  # Try 50-100 epochs for high quality
    lr=0.0001
)

# Plot loss
plt.figure(figsize=(10, 4))
plt.plot(losses, marker='o')
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
print("\nGenerating final samples...")
sample_images(model, sigmas, n_samples=16, epoch="Final")

# Show progression
samples_prog = sample_langevin_dynamics(model, sigmas, n_samples=1, n_steps_per_sigma=100)

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, (ax, sample) in enumerate(zip(axes.flatten(), samples_prog)):
    img = (sample[0] + 1) / 2
    img = img.permute(1, 2, 0).cpu().numpy()
    img = np.clip(img, 0, 1)
    ax.imshow(img)
    ax.set_title(f"σ={sigmas[i].item():.3f}")
    ax.axis('off')
plt.suptitle("Face Generation Progression")
plt.tight_layout()
plt.show()

# Save model
torch.save(model.state_dict(), 'celeba_score_model.pth')
print("\n✅ Model saved to 'celeba_score_model.pth'")
print("Training complete! Increase epochs to 50+ for photorealistic results.")
```
